```python
split_time = 350

train = data2.loc[:split_time, [ 'Robot_LeftFoot.Z']]
test = data2.loc[split_time:, [ 'Robot_LeftFoot.Z']]

ax = train.plot()
test.plot(ax=ax)
plt.legend(['train','test'])
```

```python
print(train.head())
print(test.head())
```

```python
for s in range(1, 13):
    train['shift_{}'.format(s)] = train['Robot_LeftFoot.Z'].shift(s)
    test['shift_{}'.format(s)] = test['Robot_LeftFoot.Z'].shift(s)

train.head(13)
```

```python
X_train = train.dropna().drop('Robot_LeftFoot.Z', axis=1)
y_train = train.dropna()[['Robot_LeftFoot.Z']]

X_test = test.dropna().drop('Robot_LeftFoot.Z', axis=1)
y_test = test.dropna()[['Robot_LeftFoot.Z']]

X_train.head()
```

```python
print(type(X_train))
X_train = X_train.values
print(type(X_train))
X_test = X_test.values

y_train = y_train.values
y_test = y_test.values
```

```python
print(X_train.shape)
print(y_train.shape)
```

```python
X_train_t = X_train.reshape(X_train.shape[0], 12, 1)
X_test_t = X_test.reshape(X_test.shape[0], 12, 1)

print("최종 DATA")
print(X_train_t.shape)
print(X_train_t)
print(y_train)
```

```python
from keras.layers import LSTM
from keras.models import Sequential
from keras.layers import Dense
import keras.backend as K
from keras.callbacks import EarlyStopping

K.clear_session()

model = Sequential()
model.add(LSTM(20, input_shape=(12, 1)))
model.add(Dense(1))
model.compile(loss='mean_squared_error', optimizer='adam')
model.summary()
```

```python
early_stop = EarlyStopping(monitor='loss', patience=1, verbose=1)

model.fit(X_train_t, y_train, epochs=100,
          batch_size=30, verbose=1, callbacks=[early_stop])
```

```python
y_pred = model.predict(X_test_t)
print(y_pred)
```

```python
print(type(y_pred))
plt.plot(y_pred)
```

```python
pred = pd.DataFrame(y_pred, columns=['Robot_LeftFoot.Z'])
pred.plot()
```
