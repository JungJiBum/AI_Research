```python
RF = ['Robot_RightFoot.X', 'Robot_RightFoot.Y', 'Robot_RightFoot.Z']
RF_features = walk[RF]
RFF = RF_features.plot(subplots=True)
```

```python
RF2 = ['Robot_RightFoot.X', 'Robot_RightFoot.Y', 'Robot_RightFoot.Z']
RF_features2 = walk2[RF2]
RFF2 = RF_features2.plot(subplots=True)
```

```python
train = RF_features
test = RF_features2
```

```python
train_df = pd.DataFrame(train, columns=['Robot_RightFoot.X'], index=train.index)
test_df = pd.DataFrame(test, columns=['Robot_RightFoot.X'], index=test.index)
```

```python
print(train_df)
```

```python
for i in range(1 , 13):
    train_df['shift_{}'.format(i)] = train_df['Robot_RightFoot.X'].shift(i)
    test_df['shift_{}'.format(i)] = test_df['Robot_RightFoot.X'].shift(i)
    
train_df
# train_df.tail(5)
# plt.plot(train_df)
```

```python
X_train = train_df.dropna().drop('Robot_RightFoot.X', axis=1)
Y_train = train_df.dropna()[['Robot_RightFoot.X']]

X_test = test_df.dropna().drop('Robot_RightFoot.X', axis=1)
Y_test = test_df.dropna()[['Robot_RightFoot.X']]

X_train
```

```python
print(type(X_train))
X_train = X_train.values
print(type(X_train))
X_test= X_test.values
Y_train = Y_train.values
Y_test = Y_test.values
```

```python
print(X_train.shape[0])
print(Y_train.shape)
```

```python
X_train_t = X_train.reshape(X_train.shape[0], 12, 1)
X_test_t = X_test.reshape(X_test.shape[0], 12, 1)

print("최종 DATA")
print(X_train_t.shape)
print(X_train_t)
print(Y_train)
```

```python
from keras.layers import LSTM 
from keras.models import Sequential 
from keras.layers import Dense 
import keras.backend as K 
from keras.callbacks import EarlyStopping
from keras.layers import Input

K.clear_session()
    
model = Sequential() # Sequeatial Model 
# model.add(Input(shape=(12,1)))
model.add(LSTM(20, input_shape=(12,1))) # (timestep, feature) 
model.add(Dense(1)) # output = 1 
model.compile(loss='mean_squared_error', optimizer='adam') 
model.summary()
```

```python
early_stop = EarlyStopping(monitor='loss', patience=1, verbose=1)

model.fit(X_train_t, Y_train, epochs=100,
          batch_size=30, verbose=1, callbacks=[early_stop])
```

```python
Y_pred = model.predict(X_test_t)
# print(Y_pred)
Y_pred.shape
```

```python
print(type(Y_pred))
plt.plot(Y_pred)
```
