```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

walk = pd.read_csv('walk_1.csv')
walk
```

```python
walk = walk.set_index('time')
walk
```

```python
RF = ['RightFoot.x','RightFoot.y','RightFoot.z']
plot_features = walk[RF]
_ = plot_features.plot(subplots=True)
```

```python
walk.describe().transpose()
```

```python
split_time = 6.0

train = walk.loc[:split_time, ['RightFoot.x']]
test = walk.loc[split_time:, ['RightFoot.x']]

ax = train.plot()
test.plot(ax=ax)
plt.legend(['train','test'])
```

```python
train_df = pd.DataFrame(train, columns=['RightFoot.x'], index=train.index)
test_df = pd.DataFrame(test, columns=['RightFoot.x'], index=test.index)
print(train_df.head())
print(test_df.head())
```

```python
for i in range(1,13):
    train_df['shift_{}'.format(i)] = train_df['RightFoot.x'].shift(i)
    test_df['shift_{}'.format(i)] = test_df['RightFoot.x'].shift(i)
    
train_df.head(13)
```

```python
X_train = train_df.dropna().drop('RightFoot.x', axis=1)
Y_train = train_df.dropna()[['RightFoot.x']]

X_test = test_df.dropna().drop('RightFoot.x', axis=1)
Y_test = test_df.dropna()[['RightFoot.x']]

X_train.head()
```

```python
print(type(X_train))
X_train = X_train.values
print(type(X_train))
X_test= X_test.values
Y_train = Y_train.values
Y_test = Y_test.values
```

```python
print(X_train.shape)
print(Y_train.shape)
```

```python
X_train_t = X_train.reshape(X_train.shape[0], 12, 1)
X_test_t = X_test.reshape(X_test.shape[0], 12, 1)

print("최종 DATA")
print(X_train_t.shape)
print(X_train_t)
print(Y_train)
```

```python
from keras.layers import LSTM 
from keras.models import Sequential 
from keras.layers import Dense 
import keras.backend as K 
from keras.callbacks import EarlyStopping
from keras.layers import Input

K.clear_session()
    
model = Sequential() # Sequeatial Model 
# model.add(Input(shape=(12,1)))
model.add(LSTM(20, input_shape=(12,1))) # (timestep, feature) 
model.add(Dense(1)) # output = 1 
model.compile(loss='mean_squared_error', optimizer='adam') 
model.summary()
```

```python
early_stop = EarlyStopping(monitor='loss', patience=1, verbose=1)

model.fit(X_train_t, Y_train, epochs=100,
          batch_size=30, verbose=1, callbacks=[early_stop])
```

```python
Y_pred = model.predict(X_test_t)
print(Y_pred)
```
