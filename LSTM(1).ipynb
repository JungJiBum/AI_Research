```python
split_date = 7.0
split_date = str(split_date)

train = walk.loc[:split_date, ['RightFoot.x']]
test = walk.loc[split_date:, ['RightFoot.x']]

ax = train.plot()
test.plot(ax=ax)
plt.legend(['train', 'test'])
```

```python
from sklearn.preprocessing import MinMaxScaler

sc = MinMaxScaler()

train_sc = sc.fit_transform(train)
test_sc = sc.fit_transform(test)

# train_sc
```

```python
train_sc_df = pd.DataFrame(train_sc, columns=['RightFoot.x'], index=train.index)
test_sc_df = pd.DataFrame(test_sc, columns=['RightFoot.x'], index=test.index)
train_sc_df.head()
test_sc_df.head()
```

```python
for s in range(1, 13):
    train_sc_df['shift_{}'.format(s)] = train_sc_df['RightFoot.x'].shift(s)
    test_sc_df['shift_{}'.format(s)] = test_sc_df['RightFoot.x'].shift(s)

train_sc_df.head(13)
```

```python
X_train = train_sc_df.dropna().drop('RightFoot.x', axis=1)
y_train = train_sc_df.dropna()[['RightFoot.x']]

X_test = test_sc_df.dropna().drop('RightFoot.x', axis=1)
y_test = test_sc_df.dropna()[['RightFoot.x']]

X_train.head()
```

```python
print(type(X_train))
X_train = X_train.values
print(type(X_train))
X_test= X_test.values
y_train = y_train.values
y_test = y_test.values
 

print(X_train.shape)
print(y_train.shape)
```

```python
X_train_t = X_train.reshape(X_train.shape[0], 12, 1)
X_test_t = X_test.reshape(X_test.shape[0], 12, 1)

print("최종 DATA")
print(X_train_t.shape)
print(X_train_t)
print(y_train)
```

```python
from keras.layers import LSTM 
from keras.models import Sequential 
from keras.layers import Dense 
import keras.backend as K 
from keras.callbacks import EarlyStopping
from keras.layers import Input

K.clear_session()
    
model = Sequential() # Sequeatial Model 
# model.add(Input(shape=(12,1)))
model.add(LSTM(20, input_shape=(12,1))) # (timestep, feature) 
model.add(Dense(1)) # output = 1 
model.compile(loss='mean_squared_error', optimizer='adam') 
model.summary()
```

```python
early_stop = EarlyStopping(monitor='loss', patience=1, verbose=1)

model.fit(X_train_t, y_train, epochs=100,
          batch_size=30, verbose=1, callbacks=[early_stop])
```

```python
y_pred = model.predict(X_test_t)
print(y_pred)
```
